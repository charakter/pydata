表形式のデータは[pandas.io.parsers.read_table()](http://pandas.pydata.org/pandas-docs/version/0.13.1/generated/pandas.io.parsers.read_table.html)を使って読み込みます．  
pandas.io.parsers.read_table()は任意のデリミタで区切られたファイルをDataFrameに読み込みます．  
それぞれのデータのレコードは「::」を区切り記号として使用されているので，「::」を区切り記号にして分割します

```python
import pandas as pd

# users.dat を読み込む
unames = ['user_id', 'gender', 'age', 'occupation', 'zip'] # ユーザID，性別，年齢，職業，郵便番号
users = pd.read_table('pydata-book/ch02/movielens/users.dat', sep='::', header=None, names=unames)

# ratings.dat を読み込む
rnames = ['user_id', 'movie_id', 'rating', 'timestamp'] # ユーザID，映画ID，評価，記録日時
ratings = pd.read_table('pydata-book/ch02/movielens/ratings.dat', sep='::', header=None, names=rnames)

# movies.dat を読み込む
mnames = ['movie_id', 'title', 'genres'] # 映画ID，タイトル，ジャンル
movies = pd.read_table('pydata-book/ch02/movielens/movies.dat', sep='::', header=None, names=mnames)
```

読み込んだそれぞれのデータフレームの中身を確認します．  
Pythonのスライス表現を使って先頭の5行分を表示します．

```python
# ユーザ固有の情報（ユーザID，性別，年齢，職業，郵便番号）
users[:5]
```

```python
# 映画評価の情報（ユーザID，映画ID，評価，記録日時）
ratings[:5]
```

```python
# 映画固有の情報（映画ID，タイトル，ジャンル）
movies[:5]
```

年齢は10歳刻みに分けられたグループ化を符号化した値，職業もそれぞれをカテゴリ分けし符号化した値となっています．

```python
ratings
```

[pandas.DataFrame.merge()](http://pandas.pydata.org/pandas-docs/version/0.13.1/generated/pandas.DataFrame.merge.html)を使って，3つの[データフレームオブジェクトを1つにまとめます](http://pandas.pydata.org/pandas-docs/stable/merging.html)．

1. まず評価情報表(ratings)とユーザ情報表(users)とを結合
2. 次にその結果と映画情報表(movies)を結合

pandas.DataFrame.merge()は各列の名前をキーとして共通情報を推定します．

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data
```

上記表の1項目目の情報を切り出します．

```python
data.ix[0]
```

正直，通常のインデックス指定による参照とixフィールドによる参照の違いがよくわかってないです（更に言うとindexフィールドも）．  
恐らく，通常のインデックス参照は整数で指定する**行**と名前で指定する**列**の両方を簡易に使い分けられるようになっているのに対し，  
ixフィールドによる参照は行・列を組み合わせた，複雑なスライシングに対応するために簡易性のないインターフェースを提供するものと思います．  
＃ 興味本位でソースコードをちょっとだけ覗き込んだら… (๑꒪ㅁ꒪๑)

dataデータフレームから，ある映画について、性別・ 年齢ごとの平均評価を出してみます．  
[pandas.tools.pivot.pivot_table()](http://pandas.pydata.org/pandas-docs/version/0.13.1/generated/pandas.tools.pivot.pivot_table.html)メソッドを使ってデータフレームを要約します．pivot_table()は小計を計算するのに便利らしいです．  
titleを行とし，genderを列として，ratingの値について平均をとります．

```python
mean_ratings = data.pivot_table('rating', rows='title', cols='gender', aggfunc='mean')
```

```python
mean_ratings[:5]
```

※ トップレベル関数のpandas.pivot_table関数を使う方法もあります．

```python
mean_ratings = pd.pivot_table(data, values='rating', rows='title', cols='gender', aggfunc='mean')
```

```python
mean_ratings[:5]
```

分析対象を 250 件以上のレビューのある映画に対してのみに絞り込みます．

まず各映画タイトルに対するレビュー数を集計します．集計には[pandas.DataFrame.groupby()](http://pandas.pydata.org/pandas-docs/version/0.13.1/generated/pandas.DataFrame.groupby.html)メソッドを使います（groupby()は前節で前掲）．

```python
ratings_by_title = data.groupby('title').size()
```

```python
ratings_by_title[:10]
```

groupby

インデックス参照を使って，250以上のレビュー数の映画タイトルのみを抽出します．

```python
ratings_by_title >= 250
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
```

```python
active_titles
```

得られた250件以上のレビューが得られている映画タイトル情報active_titlesをixフィールドに渡して，それぞれの男女別平均評価値の集計を取得します．

```python
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings
```

女性ユーザ評価の高い映画タイトル上位10件を取得してみます．  
[pandas.DataFrame.sort_index()](http://pandas.pydata.org/pandas-docs/version/0.13.1/generated/pandas.DataFrame.sort_index.html)メソッドを使って，'F'を軸に降順にソートします．

```python
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
```

```python
top_female_ratings[:10]
```

##  2.2.1 評価の分かれた映画の抽出

男性評価平均と女性評価平均の差を計算して'diff'列に格納します，

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

'diff'列を軸にして昇順にソートします．

```python
sorted_by_diff = mean_ratings.sort_index(by='diff')
```

```python
# 正順に15件を抽出 → 女性には評判が高いが男性には評判が低い傾向の映画タイトルが抽出される
sorted_by_diff[:15]
```

```python
# 降順に15件を抽出 → 男性には評判が高いが女性には評判が低い傾向の映画タイトルが抽出される
sorted_by_diff[::-1][:15]
```

各評価値の分散を計算することで，男女問わず評価が大きく割れた映画タイトルを抽出します．  
元のデータフレームdataから'title'と'rating'をキーにして，映画タイトルごとの評価値の標準偏差を計算します．

```python
# 映画タイトルごとの評価値の標準偏差の計算
rating_std_by_title = data.groupby('title')['rating'].std()
```

インデックス参照を使って，250以上のレビュー数の映画タイトルのみを抽出します．

```python
# 評価件数 250 件以上の映画のみを抽出(active_titles として計算済み)
rating_std_by_title = rating_std_by_title.ix[active_titles]
```

分散値の大きな映画タイトルが，評価の割れた映画タイトルとなります．[pandas.Series.order()](http://pandas.pydata.org/pandas-docs/version/0.13.1/generated/pandas.Series.order.html)はSeriesデータをソートするのに使います．

```python
# 得られた rating_std_by_title(シリーズオブジェクト)を降順でソート
rating_std_by_title.order(ascending=False)[:10]
```
